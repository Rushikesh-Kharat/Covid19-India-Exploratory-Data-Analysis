```python
import requests
import pandas as pd

# fetch data from URL
response = requests.get("https://data.covid19india.org/v4/min/timeseries.min.json")
data = response.json()

# initialize lists to store data
state_codes = []
dates = []
confirmed = []
recovered = []
deceased = []
tested = []
vaccinated = []

# loop through data and extract relevant information
for state_code in data:
    for date in data[state_code]['dates']:
        state_codes.append(state_code)
        dates.append(date)
        confirmed.append(data[state_code]['dates'][date].get('total', {}).get('confirmed', 0))
        recovered.append(data[state_code]['dates'][date].get('total', {}).get('recovered', 0))
        deceased.append(data[state_code]['dates'][date].get('total', {}).get('deceased', 0))
        tested.append(data[state_code]['dates'][date].get('total', {}).get('tested', 0))
        vaccinated.append(data[state_code]['dates'][date].get('total', {}).get('vaccinated1', 0))

# create DataFrame
df = pd.DataFrame({
    'state_code': state_codes,
    'date': dates,
    'confirmed': confirmed,
    'recovered': recovered,
    'deceased': deceased,
    'tested': tested,
    'vaccinated': vaccinated
})
```

```python
df
```

```python
df.to_csv('timeseries.csv')
```

```python
import requests
import pandas as pd
from pandas import json_normalize

# fetch data from URL
response = requests.get("https://data.covid19india.org/v4/min/timeseries.min.json")
data = response.json()

# initialize lists to store data
state_codes = []
dates = []
others = []

# loop through data and extract relevant information
for state_code in data:
    for date in data[state_code]['dates']:
        state_codes.append(state_code)
        dates.append(date)
        others.append(data[state_code]['dates'][date])

# create DataFrame
df = pd.DataFrame({
    'state_code': state_codes,
    'date': dates
})

# normalize other columns
df = pd.concat([df, json_normalize(others)], axis=1)
```

```python
df.to_csv('timeseries2.csv')
```
